```python
from statistics import mean,median,mode,multimode
dataset_A= [35000,40000,45000,50000,55000,60000,200000]
dataset_B=[40,50,55,55,60,65,70]
dataset_C=[5,5,5,6,6,6,7,8,9,10]
dataset = {"dataset A":dataset_A,"dataset B":dataset_B,"dataset_C":dataset_C}
```

```python
for name, data in dataset.items():
    print("Name:{0} Data:{1}".format(name,data))
    mean_value=mean(data)
    median_value=median(data)
    modes=multimode(data)
    print("Mean:",mean_value)
    print("Median:",median_value)
    print("Modes:",modes)
    print("Analysis:")
    if max(data) - min(data) > 2 * mean_value:
        print("The dataset might have outliers.Median is better choice")
    elif len(modes) > 1:
        print("The dataset has frequent values.Mode is better choice")
    else:
        print("This dataset seems symmetric.Mean might be a good choice")
```
